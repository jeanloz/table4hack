```python
import numpy as np
import cv2 
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import os
import sys
```

```python
def make_square(im, min_size=128, fill_color=(255, 255, 255, 255)):
    x, y = im.size
    size = max(min_size, x, y)
    new_im = Image.new('RGBA', (size, size), fill_color)
    new_im.paste(im, (int((size - x) / 2), int((size - y) / 2)))
    return new_im
```

```python
def convert_image(img,dir,file):
    
    if not os.path.exists(converted_base_image_dir+dir):
        # if here current thread is stopped and the same dir is created in other thread
        # the next line will raise an exception
            os.makedirs(converted_base_image_dir+dir )
    im = Image.open(img)

    im=make_square(im)
    im=im.resize((128,128))
    print(im.size)
    im=im.convert('RGB')
    im=ImageOps.autocontrast(im)

    im.save(converted_base_image_dir+dir+'/'+file,'JPEG')
    print(converted_base_image_dir+dir+'/'+file)
    return im
```

```python
# base_image_url = "https://52.175.195.236:8000/user/hacker1/tree/gear_images/"
base_image_dir = "/home/hacker1/notebooks/gear_images/"
square_image_dir = "/home/hacker1/notebooks/gear_images_square/"
converted_base_image_dir = "/home/hacker1/notebooks/gear_images_converted/"

if not os.path.exists(converted_base_image_dir ):
    # if here current thread is stopped and the same dir is created in other thread
    # the next line will raise an exception
    os.makedirs(converted_base_image_dir ) 

    
# converted_base_image_url = "https://52.175.195.236:8000/user/hacker1/tree/gear_images/"

for subdir,dirs, files in os.walk(base_image_dir):
    #print(dirs)
    for dir in dirs:
        
        for file in os.listdir(base_image_dir+dir):
            convert_image(subdir+dir+'/'+file,dir,file)
        
```
